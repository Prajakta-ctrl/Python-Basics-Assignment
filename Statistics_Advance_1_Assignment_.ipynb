```python
#Que.4.
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

n = 10
p = 0.5
size = 500

data = np.random.binomial(n=n, p=p, size=size)

plt.hist(data, bins=np.arange(-0.5, n+1.5, 1), density=True, edgecolor='purple', alpha=0.7, label='Histogram')

x = np.arange(0, n+1)
pmf = binom.pmf(x, n=n, p=p)
plt.scatter(x, pmf, color='green', label='Theoretical PMF')
plt.vlines(x, 0, pmf, colors='green', linestyles='dashed')

plt.title("Binomial Distribution (n=10, p=0.5)")
plt.xlabel("Number of Successes")
plt.ylabel("Probability")
plt.legend()
plt.grid(True)

plt.show()
```

```python
#Que.5.
from scipy.stats import poisson
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(0, 200, 0.8)
y = poisson.pmf(x, mu=40, loc=10)
plt.plot(x, y)
plt.show()
```

```python
#Que.7.
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

low = 20
high = 50
size = 2000

data = np.random.uniform(low=low, high=high, size=size)
sns.histplot(data, bins=30, kde=False, color='skyblue', edgecolor='black')

plt.title(f"Uniform Distribution (Range: {low} to {high})")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

```python
#Que.9.
# Importing the libraries
import numpy as np
import scipy as scipy

# Sample dataset
dataset = np.array([10,20,3,40,70,89,98,76,55,4,34])
print(f"The dataset is {dataset}")

# Calculation of z-score for values...
z_score = scipy.stats.zscore(dataset, axis=0, ddof=0, nan_policy="propagate")
print("The z-score in the dataset is", z_score)
```
