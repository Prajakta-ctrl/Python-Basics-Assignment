```python
#Que.2.
import pandas as pd

def mean_of_column(df,column_name):

  if column_name in df.columns:
    return df[column_name].mean()

  else:
    return None

  data={"c1": [2,3,4],"c2": [5,6,7],"c3": [8,9,10]}
  df=pd.DataFrame(data)

  mean_of_column1= mean_of_column(df,"c1")
  print(f"mean_of_c1: {mean_of_column1}")

  mean_of_column2= mean_of_column(df,"c2")
  print(f"mean_of_c2: {mean_of_column2}")

  mean_of_column3= mean_of_column(df,"c3")
  print(f"mean_of_c3: {mean_of_column3}")

mean_of_column(df, "c1")
```

```python
#Que.7.
import numpy as np

def multiplication_element_wise(array1,array2):

  if len(array1)==len(array2):
    result_data=np.multiply(array1,array2)
    print(result_data)

  else:
    print("None")

array1=np.array([2,3,4,5,6])
array2=np.array([7,8,9,10,11])
result_data=multiplication_element_wise(array1,array2)
print(result_data)
```

```python
#Que.8.
import matplotlib.pyplot as plt

x = [22,3,4,5,67]
y1 = [2,44,54,23,11]
y2 = [9,8,7,12,23]
y3 = [13,31,21,7,9]
plt.plot(x, y1, label='Line 1', linestyle='-', marker='o', color='black')
plt.plot(x, y2, label='Line 2', linestyle='--', marker='x', color='magenta')
plt.plot(x, y3, label='Line 3', linestyle=':', marker='s', color='green')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Multiple Line Plot")
plt.legend()
plt.show()
```
